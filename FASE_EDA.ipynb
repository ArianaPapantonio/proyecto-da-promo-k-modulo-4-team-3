```python
# GESTIONAMOS LOS NULOS

columnas_unknown = ["ndice de igualdad legal", "legalidad_colectivo_lgbtiq+", 'fecha_despenalizacin', 'constitucin', 'proteccin_amplia', 'empleo', 'crimen_odio', 'incitacin', 'prohibicin_terapias', 'matrimonio_igualitario', 'uniones_civiles', 'adopcin_conjunta', 'adopcin_segundo_progenitor']   

# iteramos por la lista de columnas a las que le vamos a cambiar los nulos por "Uknown"
for columna in columnas_unknown:
    
    # reemplazamos los nulos por el valor Unknown para cada una de las columnas de la lista
    df_unido2[columna] = df_unido2[columna].fillna("Unknown")

print("Despus del reemplazo usando 'fillna' quedan los siguientes nulos")
df_unido2[columnas_unknown].isnull().sum()
```

```python
df_unido2[df_unido2['fecha_despenalizacin'] == 'Unknown']['pas']
```

```python
# PARA PODER PASAR LA COLUMNA A INT, LE PED A CHAT QUE ME DIJERA LA INFO DE LOS PAISES FALTANTES, PORQ NO ME DEJABA CON 'Unknown'
despenalizacion = {
    'Bosnia and Herzegovina': 1998,
    'Brunei': 2007,  
    'Central African Republic': 1996,
    'Cote d\'Ivoire': 2016,
    'Czechia': 1962,
    'Democratic Republic of Congo': 0,  # Este pas sigue teniendo la homosexualidad ilegal
    'Indonesia': 0,  # La homosexualidad no est penalizada a nivel nacional, pero depende de la regin
    'Iraq': 0,  # Homosexualidad sigue siendo ilegal en Iraq
    'Saint Kitts and Nevis': 1998,
    'Saint Vincent and the Grenadines': 2001,
    'Sao Tome and Principe': 2012,
    'Taiwan': 1997
}

# Reemplazar los valores 'Unknown' en 'fecha_despenalizacin' con los aos del diccionario
for pais, ano in despenalizacion.items():
    df_unido2.loc[(df_unido2['pas'] == pais) & (df_unido2['fecha_despenalizacin'] == 'Unknown'), 'fecha_despenalizacin'] = ano
```
